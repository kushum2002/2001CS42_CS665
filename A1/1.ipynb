```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset from a CSV file
dataset_ETTh1 = pd.read_csv('ETTh1.csv')
dataset_ETTh2 = pd.read_csv('ETTh2.csv')
dataset_ETTm1 = pd.read_csv('ETTm1.csv')
dataset_ETTm2 = pd.read_csv('ETTm2.csv')
```

```python
# Date column name
date_column_name = 'date'
```

```python
# Parse the date column as a datetime object instead of string
dataset_ETTh1[date_column_name] = pd.to_datetime(dataset_ETTh1[date_column_name])
dataset_ETTh2[date_column_name] = pd.to_datetime(dataset_ETTh2[date_column_name])
dataset_ETTm1[date_column_name] = pd.to_datetime(dataset_ETTm1[date_column_name])
dataset_ETTm2[date_column_name] = pd.to_datetime(dataset_ETTm2[date_column_name])
```

```python
dataset_ETTh1.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(dataset_ETTh1['date'], dataset_ETTh1['HUFL'], label='HUFL')
# plt.plot(dataset_ETTh1['date'], dataset_ETTh1['HULL'], label='HULL')
# plt.plot(dataset_ETTh1['date'], dataset_ETTh1['MUFL'], label='MUFL')
plt.plot(dataset_ETTh1['date'], dataset_ETTh1['MULL'], label='MULL')
# plt.plot(dataset_ETTh1['date'], dataset_ETTh1['LUFL'], label='LUFL')
# plt.plot(dataset_ETTh1['date'], dataset_ETTh1['LULL'], label='LULL')
plt.plot(dataset_ETTh1['date'], dataset_ETTh1['OT'], label='OT')

plt.legend()
plt.title("Dataset Visualization for ETTh1 dataset")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.show()
```
